```python
import pandas as pd
import torch
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from PIL import Image
import open_clip
import random
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Load the BiomedCLIP model and preprocess functions
model, preprocess_train, preprocess_val = open_clip.create_model_and_transforms('hf-hub:microsoft/BiomedCLIP-PubMedBERT_256-vit_base_patch16_224')
device = torch.device('cuda') if torch.cuda.is_available() else torch.device('cpu')
model.to(device)

# Load tokenizer
tokenizer = open_clip.get_tokenizer('hf-hub:microsoft/BiomedCLIP-PubMedBERT_256-vit_base_patch16_224')
```

```python
# Load data
test_caption_file = "../Datasets/ROCO2/test_captions.csv"
test_image_folder = "../Datasets/ROCO2/test_images/test/"
test_df = pd.read_csv(test_caption_file, sep=',', header=1, names=['ID', 'Caption'])
test_df['Image'] = test_df['ID'] + ".jpg"
test_df.drop('ID', axis=1, inplace=True)

# Sample 10 random indices
num_samples = 10
random_indices = random.sample(range(len(test_df)), num_samples)
test_df_subset = test_df.iloc[random_indices].reset_index(drop=True)
print(test_df_subset )
```

```python
# Compute image embeddings
def get_image_features(image_path):
    image = Image.open(image_path).convert('RGB')
    image_tensor = preprocess_val(image).unsqueeze(0).to(device)
    with torch.no_grad():
        return model.encode_image(image_tensor).cpu().numpy()

image_features_list = []
image_paths = [test_image_folder + img_name for img_name in test_df_subset['Image']]

for img_path in image_paths:
    image_features = get_image_features(img_path)
    image_features_list.append(image_features)

# Convert list to numpy array
image_features_array = np.vstack(image_features_list)

# Compute cosine similarity matrix
cosine_sim_matrix = cosine_similarity(image_features_array, image_features_array)

# Prepare image labels
image_labels = test_df_subset['Image'].tolist()
```

```python
# Plot the heatmap
def plot_heatmap(cosine_sim_matrix, image_labels, title="Cosine similarity between images"):
    plt.figure(figsize=(12, 8))
    sns.heatmap(cosine_sim_matrix, annot=True, cmap='viridis', xticklabels=image_labels, yticklabels=image_labels)
    plt.title(title)
    plt.xlabel('Images')
    plt.ylabel('Images')
    plt.show()

# Plot the heatmap
plot_heatmap(cosine_sim_matrix, image_labels)
```
